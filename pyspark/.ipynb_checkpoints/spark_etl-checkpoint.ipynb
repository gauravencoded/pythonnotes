```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, skewness, kurtosis
spark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
df = spark.read.csv("/Users/gauravsingh/Desktop/data_science/data/ASS.csv",header=True);
#df.collect()
```

```python

print(df.count())
print(df.columns)
print(df.printSchema())
newdf=df.selectExpr("customer_id as cust_id", "gender as sex", "date as order_date", "value as order_value", "predicted_gender as predicted_sex")
print(newdf.printSchema())
print(df.dtypes)
print(df.head(1))
print(df.first())
print(df.show(6))
```

```python
df.select('value').describe().show()
df.select(skewness('value'),kurtosis('predicted_gender')).show()
df.groupBy("customer_id").agg({"value": "avg", "gender":"sum" , "date": "max", "predicted_gender":"count"}).show()
new_df=df["customer_id","value"]
new_df.show()
```
