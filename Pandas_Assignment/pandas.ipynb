```python
import pandas as  pd
df=pd.read_csv('/Users/gauravsingh/Desktop/pyspark/data/ASS.csv')
print ("Total Entries : " + str(df['customer_id'].count()))
print ("Unique Entries : " + str(df['customer_id'].nunique()))

order_counted= df.groupby('customer_id')['value'].agg('count')
most_recent_ordered=df.groupby('customer_id')['date'].agg('max')

gender_all=df[['customer_id','gender']]
c_g_d= pd.merge(gender_all, most_recent_ordered, on='customer_id', how='inner')
final=pd.merge(c_g_d,order_counted,on='customer_id')
final.columns=['customer_id','gender','most_recent_order_date','order_count']
newgrp=final.groupby('customer_id')
#res=newgrp.sort_values("customer_id", axis = 0, ascending = True, inplace = True, na_position ='last')

newgrp.first()
```
