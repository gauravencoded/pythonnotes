```python
import pandas as  pd
df=pd.read_csv('/Users/gauravsingh/Desktop/pyspark/data/ASS.csv')
# total order
order_total=df.groupby('customer_id')["value"].count().rename("order_count")
order_total
# customer if with most recent order date
recent_date=df.groupby('customer_id')['date'].max().rename('most_recent_order_date')
# removes duplicate customer_id leaving gender and customer only
gender= df.drop_duplicates(['customer_id'])[['customer_id','gender']]
#merge all data
gender_date=pd.merge(gender, recent_date, on='customer_id', how='inner')
gender_date_order=pd.merge(gender_date, order_total, on='customer_id', how='inner')
res=gender_date_order.sort_values(by=['customer_id'])
res
```
